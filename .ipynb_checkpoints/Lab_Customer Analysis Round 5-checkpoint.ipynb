```python
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(X_num)
x_normalized = transformer.transform(X_num)
print(x_normalized.shape)
# technique 1 = normalizer (pd.DataFrame(x_normalized))
#only for the numerical columns
```

```python
X_cat = data.select_dtypes(include = np.object)
X_cat
X_cat.columns
```

```python
X_cat = X_cat.drop(['customer', 'state', 'response', 'coverage', 'education',
       'employmentstatus', 'location_code', 'marital_status',
       'policy_type', 'policy', 'renew_offer_type', 'sales_channel',
       'vehicle_class', 'vehicle_size'], axis=1)
```

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(X_cat)
# encoder.categories_
encoded = encoder.transform(X_cat).toarray()
encoded
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(X_cat)
le.transform(X_cat) 
```

```python
X = np.concatenate((x_normalized, encoded), axis=1)
Y = Ydata
```

```python
X_cat.shape
```

```python
Y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=100)
```

```python
#apply the machine learn model 
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
predictions  = lm.predict(X_test)
r2_score(y_test, predictions)
#use r2 score to see if this is closer to the data (ie better at predicating 
#target_d, than without the help of the learning process)
```

```python
#some evaluation metrics

mse = mean_squared_error(y_test, predictions)
print(mse)
```

```python
rmse = math.sqrt(mse)
print(rmse)
```
